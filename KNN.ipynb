```python
import csv
import random
import math
import operator
def loadDataset(fn,split,trainingset=[],testset=[]):
  with open('/content/Iris.csv','r') as csvfile:
    lines=csv.reader(csvfile)
    dataset=list(lines)
    for x in range(1,len(dataset)-1):
      for y in range(5):
        dataset[x][y]=float(dataset[x][y])
      if random.random() < split:
        trainingset.append(dataset[x])
      else:
        testset.append(dataset[x])
  print(dataset)
```

```python
#euclidean distance
def ed(instance1,instance2,length):
  distance=0
  for x in range(1,length):
    distance+=pow((instance1[x]-instance2[x]),2)
  return math.sqrt(distance)
def getNeghibours(trainingset,testinstance,k):
  distances=[]
  print(testinstance,"\nlen",len(testinstance))
  length=len(testinstance)-1
  for x in range(len(trainingset)):
    dist=ed(testinstance,trainingset[x],length)
    distances.append((trainingset[x],dist))
  distances.sort(key=operator.itemgetter(1))
  neighbours=[]
  for x in range(k):
    neighbours.append(distances[x][0])
  return neighbours
```

```python
def getResponse(neighbours):
  classvotes={}
  for x in range(len(neighbours)):
    response=neighbours[x][-1]
    if response in classvotes:
      classvotes[response]+=1
    else:
      classvotes[response]=1
  sortedvotes=sorted(classvotes.items(),key=operator.itemgetter(1),reverse=True)
  return sortedvotes[0][0]
```

```python
def getAccuracy(testset,predictions):
  correct=0
  for x in range(len(testset)):
    if testset[x][-1]==predictions[x]:
      correct+=1
    else:
      print("Not Matched--",testset[x][-1],predictions[x])
  print("total--samples considered",len(testset))
  return (correct/float(len(testset)))*100.0
```

```python
def main():
  trainingset=[]
  testset=[]
  split=0.70
  loadDataset('/content/Iris.csv',split,trainingset,testset)
  print('train set: '+repr(len(trainingset)))
  print('test set: '+repr(len(testset)))
  #generate predictions
  predictions=[]
  k=4
  for x in range(len(testset)):
    neighbours=getNeghibours(trainingset,testset[x],k)
    print("Neighbors of ",x,testset[x]," are ",neighbours)
    result=getResponse(neighbours)
    predictions.append(result)
  accuracy=getAccuracy(testset,predictions)
  print('Accuracy: '+repr(accuracy)+'%')
main()
```
